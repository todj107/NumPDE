__Answer:__

For (i):
Using the formula $$u(x+H) = u(x) + Hu'(x)+ \frac{H^2}{2} u''(x) + \frac{H^3}{6} u'''(x),$$ where all terms of order 4 are omitted because $u \in C^3([0,1])$, we rewrite all expressions for the differences using the adequate value of $H$ ($h$ or $-h$ in this case). For the central difference, this method yields:
$$\frac{u(x+h)- u(x-h)}{2h} = \frac{((u(x) + hu'(x)+ \frac{h^2}{2} u''(x) + \frac{h^3}{6} u'''(x)))-(u(x) - hu'(x)+ \frac{h^2}{2} u''(x) - \frac{h^3}{6} u'''(x))}{2h}$$
which simplifies to 

$$\frac{1}{2h}(2hu'(x)+2\frac{h^3}{6}u'''(x))=u'(x) + \frac{h^2}{6}u'''(x).$$ Finally, subtracting $$u'(x)$$ gives: 
$$\partial^0u(x) - u'(x) = u'(x) + \frac{h^2}{6}u'''(x) - u'(x) = \frac{h^2}{6}u'''(x)$$

For (ii):
$$
    \partial^+ u =  \frac{u(x+h) - u(x)}{h} = \frac{1}{h} \left(u(x) + hu'(x) + \frac{h^2}{2}u''(x) + \frac{h^3}{6}u'''(x) - u(x) \right) = 
    u'(x) + \frac{h}{2}u''(x) + \frac{h^2}{6}u'''(x)
$$
Thus,
$$
    \partial^+ u - u'(x) = \frac{h}{2}u''(x) + \frac{h^2}{6}u'''(x)
$$

For (iii):
$$
    \partial^- u =  \frac{u(x) - u(x-h)}{h} = \frac{1}{h} \left( u(x) - u(x) + hu'(x) - \frac{h^2}{2}u''(x) + \frac{h^3}{6}u'''(x) \right) =

$$
Thus,
$$

## Problem 3

The goal of this problem is to investigate the numerical error introduced by the FDM more quantitatively and to familiarize us with the **method of manufactured solution**.

The idea is to assess the accurracy and correctness of a PDE solver implementation by constructing a know reference solution which solves the PDE problem at hand. This can be simply done by picking a meaningful and not too boring analytical solution
and explicitly calculate the data which need to be supplied, e.g.,
the right-hand side or boundary values for various boundary condition.

For instance, taking the function $u(x) = x + \cos(2 \pi x)$, we can simply calculate that
\begin{align}\notag
 u'(x) &= 1 - 2\pi \sin(2\pi x)
 \\ \notag
 u''(x) &= - (2\pi)^2 \cos(2\pi x)
\end{align}
and thus $u$ satisfies the Poisson problem
$$
-u''(x) = (2\pi)^2 \cos(2\pi x)
$$
with boundary conditions
$$
u'(0) = 1, \quad u(1) = 2.
$$

With a known reference solution at hand we can compute
the error vector $u_i - U_i$ at the grid points $\{x_i\}_{i=0}^N$
for a series of successively refined grid, e.g. by taking
$N = 4\cdot2^k$ for $k = 0, 1, 2, 3$ etc.

Reducing the mesh size $h$ by half allows us to easily compute
the **experimental order of convergence (EOC)**,
that is the observed error reduction in the numerical solution
when passing from a coarser mesh with mesh size $h$ to
to a finer mesh with $h/2$.
The EOC can then be compare with the theoretically predicted error reduction (if known).

For instance, if you know that the discretization error $E(h)$ given on mesh
with mesh size $h$ and measured in some norm $ \| \cdot \| $ behaves like 
$\| E(h) \| \sim h^q$, you can conclude that  
$$
\dfrac{\|E(h/2)\|}{\|E(h)\|} \sim \dfrac{(h/2)^q}{h^q} = (1/2)^q
$$
when passing from $h$ to half the mesh size $h/2$. Taking the logarithm
of the last equation shows 
$$
q \sim \dfrac{\ln\bigl(\|E(h)\|/\|E(h/2)\|\bigr)}{\ln 2}. 
$$
(Verify this!)
Thus the EOC for each mesh size h is measured by the 
$$
EOC(h) = \dfrac{\ln\bigl(\|E(h)\|/\|E(h/2)\|\bigr)}{\ln 2}.
$$

Alternatively, you can do a log-log plot of your error as function of $h$,
that is plot $\ln (\| E(h)\|)$ againts $\ln(h)$. Then the slope of this
plot should be $\sim q$ if we expect the method to be of convergence order $q$.

```python
# Array and stuff 
import numpy as np
# Linear algebra solvers from scipy
import scipy.linalg as la
# Basic plotting routines from the matplotlib library 
import matplotlib.pyplot as plt
import pandas as pd

def work_func_FDM(bc):
    N_list = [4, 8, 16, 32, 64]
    E_norm = np.zeros(len(N_list))
    h_i = np.zeros(len(N_list))
    
    j = 0
    for N in N_list:
    	# Initial values
        x = np.linspace(0,1,N+1)
        A = np.zeros((N+1, N+1))
        h = 1/N

        # Analytical function: -uxx
        F = (2*np.pi)**2*np.cos(2*np.pi*x)

        # Boundary values
        F[0] = 1
        F[N] = 2
        A[N,N] = h**2
        if bc == 'd':
            # Dirichlet
            A[0,0] = h**2
        elif bc == 'n':
            # Neumann
            A[0,0] = -h
            A[0,1] = h
        elif bc == "in":
            # Improved Nuemann
            A[0,0] = -h
            A[0,1] = h
            F[0] -= 2*h*np.pi**2

        for i in range(1, N):
            A[i, i-1] = -1
            A[i, i+1] = -1
            A[i, i] = 2
        
        # Scaling A 
        A = A / (h**2)
    
        U = la.solve(A, F)
        U_exact = x + np.cos(2*np.pi*x)
        
        E_norm[j] = np.max(np.abs(U - U_exact)) 
        h_i[j] = h
        j += 1

    return E_norm, h_i

# Dirichlet (a):
E_d, h_d = work_func_FDM('d')
q_d = np.polyfit(np.log(h_d), np.log(E_d), 1)[0]
plt.loglog(h_d, E_d)
plt.title("3a) Dirichlet")
plt.legend([f"q = {q_d:.2f}"])
plt.xlabel("h")
plt.ylabel("E(h)")
plt.show()

# Neumann (b):
E_n, h_n = work_func_FDM('n')
q_n = np.polyfit(np.log(h_n), np.log(E_n), 1)[0]

# Improved Nuemann (c):
E_i, h_i = work_func_FDM("in")
q_i = np.polyfit(np.log(h_i), np.log(E_i), 1)[0]

# Visualization
data = {"Dirichlet":q_d, "Neumann":q_n, "Imp. Neum.":q_i}

df = pd.DataFrame(data, index=["q:"])
df = df[["Dirichlet", "Neumann", "Imp. Neum."]]
df = df.round(3)

display(df)
```

Therefore, if we know the value at the left boundary, we can determine the solution at the points at higher values of $x$. $\partial_{+}$ tries to use the succeeding solution which causes instabilities, because there is no stable boundary condition at the edge where $\partial_{+}$ needs a value. 

```python
# 4 b)
### Array and stuff 
import numpy as np
# Linear algebra solvers from scipy
import scipy.linalg as la
# Basic plotting routines from the matplotlib library 
import matplotlib.pyplot as plt

N = 64
x = np.linspace(0, 1, N+1)
eps = 1

# Diffusion matrix initialized
C = np.zeros((N+1, N+1))

# Advection matrix initialized
B = np.zeros((N+1, N+1))

h = 1 / N

for i in range(1, N):
    C[i, i-1] = -1
    C[i, i+1] = -1
    C[i, i] = 2

for i in range(1, N):
    B[i, i-1] = -1
    B[i, i+1] = 1

# maybe?
F = np.ones(N+1)

# scale and add matrices
B = B/(2*h)
C = eps * C/(h**2)
A = B + C

# Dirichlet boundary conditions
A[0, 0] = 1
F[0] = 0

A[N, N] = 1
F[N] = 0

U = la.solve(A, F)

# "x-r" means mark data points as "x", connect them by a line and use red color
plt.plot(x, U, "x-r")

# exact solution
x = np.linspace(0, 1, 1000)
kappa = np.exp(-1/eps)
exactsol = x-(np.exp((x-1)/eps)-kappa)/(1-kappa)
plt.plot(x, exactsol, "b")

plt.xlabel("x")
plt.ylabel("u(x)")
plt.title("Numerical vs Exact Solution")
plt.show()
```

```python
# 4c) and b)
### Array and stuff 
import numpy as np
# Linear algebra solvers from scipy
import scipy.linalg as la
# Basic plotting routines from the matplotlib library 
import matplotlib.pyplot as plt
import pandas as pd

def work_func_FDM(eps, N, plot_idx, diff, plot_status):
    x = np.linspace(0, 1, N+1)

    # Diffusion matrix initialized
    C = np.zeros((N+1, N+1))

    # Advection matrix initialized
    B = np.zeros((N+1, N+1))

    h = 1/N

    for i in range(1, N):
        C[i, i-1] = -1
        C[i, i+1] = -1
        C[i, i] = 2
    if diff == 'center':
        for i in range(1, N):
            B[i, i-1] = -1
            B[i, i+1] = 1
        B = B/(2*h)
    elif diff == 'forward':
        for i in range(1, N):
            B[i, i] = -1
            B[i, i+1] = 1
        B = B/(h)
    elif diff == 'backward':
        for i in range(1, N):
            B[i, i] = 1
            B[i, i-1] = -1
        B = B/(h)
        
        
    # maybe?
    F = np.ones(N+1)

    # scale and add matrices
    
    C = eps * C/(h**2)
    A = B + C

    # Dirichlet boundary conditions
    A[0, 0] = 1
    F[0] = 0

    A[N, N] = 1
    F[N] = 0

    U = la.solve(A, F)
    kappa = np.exp(-1/eps)
    #exact solution discrete
    exactsol = x-(np.exp((x-1)/eps)-kappa)/(1-kappa)
    #Error
    E = np.max(np.abs(U-exactsol))
    #Exact solution for analytical x, needed to plot
    
    #plot-color
    colors = ['g', 'y', 'c', 'k']
    color = colors[plot_idx]
    if plot_status == 'plot':
        plt.plot(x, U, marker = 'x', color = color)
        plt.xlabel("x")
        plt.ylabel("u(x)")
    return E

#One error for the plot and one for the table
E_center = np.zeros((3,4))
E_center_tab = np.zeros((3,4))
E_frwd = np.zeros((3,4))
E_frwd_tab = np.zeros((3,4))
E_back = np.zeros((3,4))
E_back_tab = np.zeros((3,4))

eps_arr = [0.1, 0.01, 0.001]
N_arr = np.array([8, 16, 32, 64])

#######Central finite difference ########
for j in range(0,len(eps_arr)):
    eps = eps_arr[j]
    
    #Plot exact solution
    x_anal = np.linspace(0, 1, 1000)
    kappa = np.exp(-1/eps)
    exactsol_anal = x_anal-(np.exp((x_anal-1)/eps)-kappa)/(1-kappa)
    plt.figure(j)
    plt.plot(x_anal,exactsol_anal,"b")
    
    #Solve system for different N and plotting each sol in figure corresponding to j:th epsilon (plotting done in function)
    for i in range(0,len(N_arr)): 
        E_center[j,i] = work_func_FDM(eps, N_arr[i],i,'center','plot')
        #Multiply N by four to get better values for EOC determination
        E_center_tab[j,i] = work_func_FDM(eps, 10*N_arr[i],i,'center','')
    plt.legend(['Exact solution','N = 8', 'N = 16', 'N = 32 ', 'N = 64'])
    plt.title(f"$\\epsilon$ = {eps} for central finite difference")
    

#change N_arr to match N:s used in table
N_arr_cent = 10*N_arr


# Table for central
data = {"Error for eps = 0.1":E_center_tab[0,:],"Error for eps = 0.01":E_center_tab[1,:], "Error for eps = 0.001":E_center_tab[2,:]}

df = pd.DataFrame(data, index=N_arr_cent)
df = df[["Error for eps = 0.1", "Error for eps = 0.01", "Error for eps = 0.001"]]
df = df.round(6)
df.index.name="N"
display(df.style.set_caption("Table 1: Errors for central difference"))

######Forward central difference######
for j in range(0,len(eps_arr)):

    eps = eps_arr[j]
    
    #Plot exact solution
    x_anal = np.linspace(0, 1, 1000)
    kappa = np.exp(-1/eps)
    exactsol_anal = x_anal-(np.exp((x_anal-1)/eps)-kappa)/(1-kappa)
    plt.figure(j+3)
    plt.plot(x_anal,exactsol_anal,"b")
    
    #Solve system for different N and plotting each sol in figure corresponding to j:th epsilon (plotting done in function)
    for i in range(0,len(N_arr)): 
        E_frwd[j,i] = work_func_FDM(eps, N_arr[i],i,'forward','plot')
        E_frwd_tab[j,i] = work_func_FDM(eps, 100*N_arr[i],i,'forward','')
    plt.legend(['Exact solution','N = 8', 'N = 16', 'N = 32', 'N = 64'])
    plt.title(f"$\\epsilon$ = {eps} for forward finite difference")


#Table forward
N_arr_frwd = 100*N_arr
data = {"Error for eps = 0.1":E_frwd_tab[0,:],"Error for eps = 0.01":E_frwd_tab[1,:], "Error for eps = 0.001":E_frwd_tab[2,:]}

df = pd.DataFrame(data, index=N_arr_frwd)
df = df[["Error for eps = 0.1", "Error for eps = 0.01", "Error for eps = 0.001"]]
df = df.round(6)
df.index.name="N"
display(df.style.set_caption("Table 1: Errors for forward difference"))
    
##########Backward central difference############
for j in range(0,len(eps_arr)):
    eps = eps_arr[j]
    
    #Plot exact solution
    x_anal = np.linspace(0, 1, 1000)
    kappa = np.exp(-1/eps)
    exactsol_anal = x_anal-(np.exp((x_anal-1)/eps)-kappa)/(1-kappa)
    plt.figure(j+6)
    plt.plot(x_anal,exactsol_anal,"b")
    
    #Solve system for different N and plotting each sol in figure corresponding to j:th epsilon (plotting done in function)
    for i in range(0,len(N_arr)): 
        E_back[j,i] = work_func_FDM(eps, N_arr[i],i,'backward', 'plot')
        E_back_tab[j,i] = work_func_FDM(eps, 100*N_arr[i],i,'backward','')
    plt.legend(['Exact solution','N = 8', 'N = 16', 'N = 32', 'N = 64'])
    plt.title(f"$\\epsilon$ = {eps} for backward finite difference")   
  

#Table Backward
N_arr_back = 100*N_arr
data = {"Error for eps = 0.1":E_back_tab[0,:],"Error for eps = 0.01":E_back_tab[1,:], "Error for eps = 0.001":E_back_tab[2,:]}

df = pd.DataFrame(data, index=N_arr_back)
df = df[["Error for eps = 0.1", "Error for eps = 0.01", "Error for eps = 0.001"]]
df = df.round(6)
df.index.name="N"
display(df.style.set_caption("Table 1: Errors for backward difference"))

plt.show()
```
